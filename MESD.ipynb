```python
import numpy as np
import pandas as pd

from pathlib import Path
from tqdm import tqdm
import torchaudio
import librosa
import IPython.display as ipd
```

#Dataset MESD

---



```python
data = []

for path in tqdm(Path("/content/drive/MyDrive/MESD/MESDAll").glob("*_*_*.wav")):
    word = str(path).split('/')[-1].split('_')[3].split('.')[0]
    emotion = str(path).split('/')[-1].split('_')[0]
    ActorInfo = str(path).split('/')[-1].split('_')[1]
    try:
        # There are some broken files
        speech, sr = torchaudio.load(path)
        speech = speech[0].numpy().squeeze()
        speech = librosa.resample(np.asarray(speech), sr, 16_000)
        data.append({
            "Palabra": word,
            "Emoción": emotion,
            'InfoActor': ActorInfo,
            'AudioArray' : speech
        })
    except Exception as e:
        print(str(path), e)
        pass

    # break
```

```python
df = pd.DataFrame(data)
df.head()
```

##Reemplazar las etiquetas de InfoActor



```python
def limpiar_info(df):
  df['InfoActor']=df['InfoActor'].replace('C','Niño')
  df['InfoActor']=df['InfoActor'].replace('M','Hombre')
  df['InfoActor']=df['InfoActor'].replace('F','Mujer')
```

```python
limpiar_info(df)
```

##Reemplazar las etiquetas de Emoción



```python
def limpiar_emociones(df):
  df['Emoción']=df['Emoción'].replace('Anger','Enojo')
  df['Emoción']=df['Emoción'].replace('Disgust','Disgusto')
  df['Emoción']=df['Emoción'].replace('Fear','Miedo')
  df['Emoción']=df['Emoción'].replace('Happiness','Felicidad')
  df['Emoción']=df['Emoción'].replace('Sadness','Tristeza')
```

```python
limpiar_emociones(df)
```
